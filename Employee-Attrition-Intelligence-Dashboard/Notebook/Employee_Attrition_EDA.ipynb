```python
Q1 = df["MonthlyIncome"].quantile(0.25)
Q3 = df["MonthlyIncome"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df["MonthlyIncome"] < lower) |
    (df["MonthlyIncome"] > upper)
]

print(outliers)
```

```python
department_percent = pd.crosstab(
    df["Department"],
    df["Attrition"],
    normalize="index"
) * 100

department_percent
```

```python
import matplotlib.pyplot as plt

department_percent["Yes"].sort_values().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Department-wise Attrition Rate")
plt.ylabel("Attrition Percentage")
plt.show()
```
